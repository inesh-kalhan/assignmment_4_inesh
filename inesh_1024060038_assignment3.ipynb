# Method 1: Drop rows or columns that have a missing value

```python
#One option is to drop the rows or columns that contain a missing value.
df.dropna()
#With the default parameter values, the dropna function drops the rows that contain any
#missing value.
```

```python
#We can also choose to drop columns that have at least one missing value by using the
#axis parameter.
df.dropna(axis=1)
```

```python
#Another situation is to have a column or row that is full of missing values.
#The dropna function can be used for this as well. We just need to change the value of
#how parameter.
df.dropna(how="all")
```

```python
#Drop rows or columns based on a threshold value
#For instance, �thresh=4� means that the rows that have at least 4 non-missing values will be kept. The other ones will be dropped.
df.dropna(thresh=4)
```

```python
#Drop based on a particular subset of columns
#The subset parameter of the dropna function is used for this task. For instance,
#we can drop the rows that have a missing value in measure 1 or measure 2 columns as follows:
df.dropna(subset=["Measure_2","Measure_3"])
```

# Method 2: Fill with a constant value

```python
#We can choose a constant value to be used as a replacement for the missing values.
values={"Item":1014,"Measure_1":0}
df.fillna(value=values)
```

# Method 3: Fill with an aggregated value

```python
#Another option is to use an aggregated value such as mean, median, or mode.
df["Measure_2"].fillna(df["Measure_2"].mean())
```

# Method 4: Replace with the previous or next value

```python
#It is possible to replace the missing values in a column with the previous or next value
# in that column.
#This method might come in handy when working with time-series data.
#Consider you have a data frame that contains the daily temperature measurement
# and the temperate in one day is missing.
#The optimal solution would be to use the temperature in the next or previous day.
df.fillna(method="ffill")
```

```python
#we can limit the number of missing values replaced with this method.
#If we set the limit parameter as 1, then a missing value can only be replaced with its next value.
#The second or third following value will not be used for replacement.
df.fillna(method="bfill",limit=1)
```

The values will be selected according to the row indices and column names. For instance, if there is a missing value in the second row in the item column, the value in the same location in the new data frame will be used.

# Method 2: Using Z-score
Z- Score is also called a standard score. This value/score helps to understand that how far is the data point from the mean. And after setting up a threshold value one can utilize z score values of data points to define the outliers.

```python
# Z score
from scipy import stats
import numpy as np

z = np.abs(stats.zscore(df['ProductRelated_Duration']))
print(z)
```

Now to define an outlier threshold value is chosen which is generally 3.0. As 99.7% of the data points lie between +/- 3 standard deviation (using Gaussian Distribution approach).

```python
#threshold = 3

# Position of the outlier
print(np.where(z > 3))
```

# Method 3: Inter Quartile Range
Inter Quartile Range approach to finding the outliers is the most commonly used and most trusted approach used in the research field.

```python
# IQR
import numpy as np
Q1 = np.percentile(df['ProductRelated_Duration'], 25,
                   interpolation = 'midpoint')

Q3 = np.percentile(df['ProductRelated_Duration'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
```

To define the outlier base value is defined above and below datasets normal range namely Upper and Lower bounds, define the upper and the lower bound (1.5*IQR value is considered) :

upper = Q3 + 1.5*IQR

lower = Q1 - 1.5*IQR

```python
# Upper bound
upper = np.where(df['ProductRelated_Duration'] >= (Q3+1.5*IQR))
print(upper)
# Lower bound
lower = np.where(df['ProductRelated_Duration'] <= (Q1-1.5*IQR))
print(lower)
```

```python
#MISSING DATA HANDLING
import seaborn as sns
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
import numpy as np

# 1. Load dataset
titanic = sns.load_dataset("titanic")
print("Original dataset shape:", titanic.shape)
print("\nMissing values:\n", titanic.isnull().sum())

# --- METHOD 1: Dropping rows/columns ---
drop_df = titanic.dropna()
print("\nAfter dropping missing values:", drop_df.shape)

# --- METHOD 2: Replace with mean/median ---
fill_df = titanic.copy()
fill_df['age'] = fill_df['age'].fillna(fill_df['age'].median())  # median for skewed

# Convert 'deck' to object dtype before filling
fill_df['deck'] = fill_df['deck'].astype('object').fillna('Unknown')              # fill categorical
fill_df['embarked'] = fill_df['embarked'].fillna(fill_df['embarked'].mode()[0])
print("\nAfter filling missing values:", fill_df.isnull().sum().sum(), "missing left")

# --- METHOD 3: KNN Imputer ---
# Select only numeric columns for KNN
numeric_df = titanic.select_dtypes(include=[np.number])
imputer = KNNImputer(n_neighbors=5)
knn_imputed = imputer.fit_transform(numeric_df)
knn_df = pd.DataFrame(knn_imputed, columns=numeric_df.columns)
print("\nKNN Imputer done → no missing:", knn_df.isnull().sum().sum())

# --- METHOD 3B: Iterative Imputer (like MICE) ---
iter_imputer = IterativeImputer(max_iter=10, random_state=0)
iter_imputed = iter_imputer.fit_transform(numeric_df)
iter_df = pd.DataFrame(iter_imputed, columns=numeric_df.columns)
print("\nIterative Imputer done → no missing:", iter_df.isnull().sum().sum())
```

```python
# SCALING TECHNIQUES
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# 1. Generate synthetic dataset (mimicking housing data)
np.random.seed(42)
data = {
    "Income": np.random.normal(50000, 15000, 1000),   # Normal scale
    "Rooms": np.random.normal(5, 2, 1000),            # Small scale
    "Occupancy": np.random.exponential(2, 1000)       # Skewed with outliers
}
df = pd.DataFrame(data)

# 2. Apply Min-Max Scaling
minmax = MinMaxScaler()
df_minmax = pd.DataFrame(minmax.fit_transform(df), columns=df.columns)

# 3. Apply Standard Scaling
standard = StandardScaler()
df_standard = pd.DataFrame(standard.fit_transform(df), columns=df.columns)

# 4. Plot Original Distributions
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, col in enumerate(df.columns):
    sns.histplot(df[col], bins=50, ax=axes[i], kde=True)
    axes[i].set_title(f"Original: {col}")
plt.tight_layout()
plt.show()

# 5. Plot Min-Max Scaling
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, col in enumerate(df.columns):
    sns.histplot(df_minmax[col], bins=50, ax=axes[i], kde=True)
    axes[i].set_title(f"Min-Max: {col}")
plt.tight_layout()
plt.show()

# 6. Plot Standard Scaling
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, col in enumerate(df.columns):
    sns.histplot(df_standard[col], bins=50, ax=axes[i], kde=True)
    axes[i].set_title(f"Standard: {col}")
plt.tight_layout()
plt.show()
```

```python
#Custom Function for Normalization
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load dataset
iris = load_iris(as_frame=True)
df = iris.frame   # DataFrame with features + target

# Select only features (numeric columns)
features = df.iloc[:, :-1]   # first 4 columns
print(features.head())
def manual_minmax(col):
    return (col - col.min()) / (col.max() - col.min())

# Apply custom normalization
manual_scaled = features.apply(manual_minmax)
scaler = MinMaxScaler()
sklearn_scaled = pd.DataFrame(scaler.fit_transform(features), columns=features.columns)
print("\nManual Normalization (first 5 rows):")
print(manual_scaled.head())

print("\nSklearn MinMaxScaler (first 5 rows):")
print(sklearn_scaled.head())

# Check difference
diff = (manual_scaled - sklearn_scaled).abs().sum().sum()
print("\nTotal difference between methods:", diff)
```

```python
#Imputation with Domain Knowledge:
import seaborn as sns
import pandas as pd

# Load dataset
titanic = sns.load_dataset("titanic")

# Check missing values
print("Missing values before:\n", titanic.isnull().sum().head())

# Step 1: Look at Age distribution by class
print("\nAverage Age by Class:")
print(titanic.groupby("class")["age"].mean())

print("\nAverage Age by Sex:")
print(titanic.groupby("sex")["age"].mean())

# Step 2: Impute missing Age using groupby + transform
# (Option A: based on class)
titanic["age_imputed_class"] = titanic["age"].fillna(
    titanic.groupby("class")["age"].transform("mean")
)

# (Option B: based on sex)
titanic["age_imputed_sex"] = titanic["age"].fillna(
    titanic.groupby("sex")["age"].transform("mean")
)

# Step 3: Verify imputation
print("\nMissing values after class-based imputation:", titanic["age_imputed_class"].isnull().sum())
print("Missing values after sex-based imputation:", titanic["age_imputed_sex"].isnull().sum())

# Show a few rows
print(titanic[["age", "class", "sex", "age_imputed_class", "age_imputed_sex"]].head(10))
```

```python
#Detecting Outiers
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing

# 1. Load dataset
housing = fetch_california_housing(as_frame=True)
df = housing.frame

# Select the target column (house value)
col = "MedHouseVal"
data = df[col]

print("Basic stats:\n", data.describe())

# --- METHOD 1: Z-SCORE ---
mean_val = np.mean(data)
std_val = np.std(data)

z_scores = (data - mean_val) / std_val
z_outliers = data[np.abs(z_scores) > 3]   # threshold = 3

print("\nZ-score Outliers Detected:", len(z_outliers))

# --- METHOD 2: IQR ---
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

iqr_outliers = data[(data < lower) | (data > upper)]

print("IQR Outliers Detected:", len(iqr_outliers))

# --- Compare ---
print("\nSample Z-score outliers:\n", z_outliers.head())
print("\nSample IQR outliers:\n", iqr_outliers.head())
```

```python
#Feature Engineering
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# 1. Load Titanic dataset
titanic = sns.load_dataset("titanic")

# 2. Create new feature: FamilySize = SibSp + Parch + 1
titanic["FamilySize"] = titanic["sibsp"] + titanic["parch"] + 1

print(titanic[["sibsp", "parch", "FamilySize"]].head(10))

# 3. Normalize manually (0–1 range)
def manual_minmax(col):
    return (col - col.min()) / (col.max() - col.min())

titanic["FamilySize_norm"] = manual_minmax(titanic["FamilySize"])

# 4. Scale using StandardScaler (mean=0, std=1)
scaler = StandardScaler()
titanic["FamilySize_std"] = scaler.fit_transform(titanic[["FamilySize"]])

# 5. Compare results
print("\nOriginal vs Normalized vs Standardized (first 10 rows):")
print(titanic[["FamilySize", "FamilySize_norm", "FamilySize_std"]].head(10))
```
